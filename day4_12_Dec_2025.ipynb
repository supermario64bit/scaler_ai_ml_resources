# **Pandas In Detail**

**Series**<br>
1D array in pandas are called Series. But we can set custom index values for series. These custom index values are called labels. And also we can set name for the series for which entity data is reffered by the series.

````python
import pandas as pd
import numpy as np
````

````python
votes = np.array([120, 240, 150])
ratings = np.array([4.1, 4.5, 3.9])
restaurants = np.array(['R1', 'R2', 'R3'])
````

````python
votes_series = pd.Series(votes, index=restaurants, name="votes") #Votes with labels and name
votes_series
````

````python
rating_series = pd.Series(ratings, index=restaurants, name="ratings")
rating_series
````

Instead of combining them seperately we can create a 2D matrix which is called dataframe using the concatenation method.



```
pd.concat([series1, series2], axis=1)

---


```



````python
pd.concat([votes_series, rating_series], axis=1)
````

**Practice Question**<br>
Find the average ratings of the restaurants in the dataset gives above?<br>
Steps:
1.   Right now ratings is in string with /5 at the end. Need to remove it.
2.   Clean the data and make it float
3.   Calculate the mean of the ratings



````python
df_new["only_ratings"] = df_new["Ratings"].str[:-2]
df_new["only_ratings"]
````

````python
df_new["only_ratings"] = df_new["only_ratings"].replace("N", np.nan) # replacing N with nan
df_new["only_ratings"] = df_new["only_ratings"].astype(float) #Converting it into float
````

````python
df_new["only_ratings"].mean() # Mean is the average ratings
````

**Replace null value**<br>
We have fillna function to fill the columns with null value to a value.


```
df["column_name"].fillna(value_to_be_filled)
```



````python
df_new["only_ratings_1"] = df_new["only_ratings"].fillna(1) # Filling the null value with 1
````

````python
df_new["only_ratings_null_means"] = df_new["only_ratings"].fillna(df_new["only_ratings"].mean()) # Filling the null value with mean. So average value wont change
````
