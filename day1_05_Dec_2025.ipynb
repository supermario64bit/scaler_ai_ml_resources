# **NumPy**<br>
A python library which supports numerical operations and datasets where we can do vectorised and broadcasted operations. This property makes it distinct from python list. The ndArray in NumPy is the important datastructure where we do most of the operations. Here we supports vectorised and broadcasted operations and also the it is contegious memmory. But python list which has similar properties is not contegious but its references are arranged in contegious manner. Unlike python numpy ndArrays only accepts same type of data. The numpy is optimised in C programmin language which is a lower level programming language and this also makes the numpy performance faster.

# **What you mean by vectorized operations?**<br>
Consider the python list, if we need to multiply all the values inside the list means we need to run a loop and access each elements and then multiply and update the new value to the index.
<br>
Instead of doing this sequential operation, we can do it in single operation on entire arrays and the operation will broadcasted to all the array members simultaneosly. And that is vectorized operation
<br>
Python list didn't support vectorized operations but NumPy array does.

````python
import numpy as np

# Food Delivery Data Exploration and analysis

# Votes for each of the restaurants
votes = np.array([ 775,  787,  918,   88,  166,  286, 2556,  324,  504,  402])

# Average cost for each of the restaurants
costs = np.array(["'800.0'" ,"'800.0'", "'800.0'", "'300.0'", "'600.0'", "'600.0'", "'600.0'", "'700.0'" ,"'550.0'", "'500.0'"])

print("Votes (Array):", votes)
print("Costs (Array):", costs)
````

# **Data Manipulation**
We have votes and cost of respected restaurants as 1D array.

1. We can convert them into a 2d array using:


```
np.array([votes, costs])
```
2. We can create an array with specified range:


```

```




````python
np.array([votes, costs])
````

````python
np.arange(1,5) # Create an array between 1 to 4 and 5 is excluded
````

````python
np.arange(1,9,2) #Create an array from 1 to 8 by skipping 2 numbers.
````

3. In our costs, the data is available in string and it cannot convert directly to float because of the single inverted comma symbol present in the string. So first we will remove the symbol and then we can convert directly to float.
<br>
To remove symbol:


```
np.char.replace(costs, "'", "")
```

To convert to float:


```
costs.astype(float)
```





````python
cleaned_costs = np.char.replace(costs, "'", "")
````

````python
cleaned_costs = cleaned_costs.astype(float)
````

# **Accesing Array Data With Index**

To get a single element in index i, we can use:
```
cleaned_costs[i]
```

To get elements from reverse order user negative indexing:
```
cleaned_costs[-i]
```
To get elements in multiple index as array, we can use:
```
cleaned_costs[1,3,4]
```
To slice the array with index:
```
cleaned_costs[1:3]
```
To filter data with some condition
```
cleaned_costs[cleaned_costs > 200]
```

````python
cleaned_costs[1] # To get second element
````

````python
cleaned_costs[-1] # To get last element
````

````python
cleaned_costs[[0,2,3]] # To get elements in the index 0,2,3 as array
````

````python
cleaned_costs[1:3] # Slice elements from 1 to 3 index, 3 is excluded
````

````python
cleaned_costs[cleaned_costs < 500] # Get elements less than 500
````
