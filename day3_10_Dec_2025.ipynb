# **Broadcasting In NumPy**
In general broadcasting means, something is broadcasted from one source to multiple listeners.<br><br>
Usually we cannot do mathematically operations like +,-, etc to the matrics in different shape.


```
[[1,2,3],[4,5,6]] + 10
```
The above expression is not compatible, because both of them are in differrent shape.But in Numpy it is possible.<br>

In numpy the elements will be broadcasted to a compatible shape and complete the operation.
```
[[1,2,3],[4,5,6]] + 10
```
The above matrix will converted to
```
[[1,2,3],[4,5,6]] + [[10,10,10],[10,10,10]]
```
Now we can perform the operation because both are in compatible shape.


````python
import numpy as np
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("array_2d: ")
print(array_2d)
print()
print("array_2d + 10 = ")
print(array_2d+10)
````

Not only single digit, we can also do the operations to two matrix with different shape. It will also broadcasted to compatable matrix and can perform the operations.

Eg:

```
[[1,2,3]] + [[4],[5],[6]]
```
The above matrix are in different shape and the broadcasting will work in the below form:
```
[[1,2,3],[1,2,3],[1,2,3]] + [[4.4.4],[5,5,5],[6,6,6]]
```

````python
np.array([[1, 2, 3]]) + np.array([[4],[5],[6]])
````

**Incompatible shape for broadcasting in NumPy**<br>
Numpy compares the shape from right to left for broadcasting. If at any position the sizes are neither equal nor 1, the broadcasting fails and that is an incompatible shape for broadcasting.



```
[[1, 2, 3], [4, 5, 6]] + [10, 20]
```
The above matrix are not compatible because their position sizes are not equal or 1. So here broadcasting is not possible and numpy will through an error.


# **Split Function In NumPy**
Split function in numpy is used to divide a single array into multiple sub arrays.

```
np.split(array_2d, 2, axis = 1)
[link text](https://)```



````python
print("Original array:")
array_2d = np.array([[1, 2, 3, 4],
                     [5, 6, 7, 8]])
print(array_2d)
print()
print("Split column wise into two:")
print(np.split(array_2d, 2, axis = 1))
print()
print("Split row wise into two:")
print(np.split(array_2d, 2, axis = 0))
````

Instead of mentioning the axis as arguments we have a lazy method to perform.

To Split column wise we can use hsplit/ horizontal split:

```
np.hsplit(array_2d, 2)
```
To Split row wise we can use vsplit/ vertical split:

```
np.vsplit(array_2d, 2)
```


````python
np.hsplit(array_2d, 2)
````

````python
np.vsplit(array_2d, 2)
````

We can use index wise split. We can mention the index where we need to split and the numpy will create subset for each arrays.
```
np.split(array, [2,5])
```
In the above expression the array will split into 3 parts, 0 to 2(exclusing 2), 2 to 5(excluding 5), 5 to end of the index


````python
x = np.array([1, 2, 3, 4, 5, 6])
np.split(x, [2,5])
````

**Stacking in NumPy**<br>
Stacking in numpy means combining multiple arrays along an existing or new axis.

```
np.stack([a, b])
```



````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Adds a new axis
np.stack([a, b])

# Stack column wise
np.hstack([a, b])

# Stack row wise
np.vstack([a, b])
````

**Concatenation in NumPy**<br>
Concatination means joins two arrays.



```
np.concatenate([array1, array2])
```



````python
np.concatenate([a, b])
````
