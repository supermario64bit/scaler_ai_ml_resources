**Filtering Data**


---


To filter the data we can use where function in NumPy array. But we have some other functions to check specific conditions. They are:



1.   np.any(condition_here)
*any* function returns boolean value. If the given condition satisfy any of the data in the array it will return true, otherwise it will return false.

For checking the all_costs has any cost which is greater than 500:
```
np.any(all_costs > 500)
```


2.   np.all(condition_here)
*all* function also returns boolean value. If the given condition is satosfied by all the datas in the array, the function will return true. Otherwise it always return false.

For checking the all_costs has all the cost which is greater than 500:
```
np.all(all_costs > 500)
```

````python
print(np.any(all_costs > 500))
````

````python
print(np.all(all_costs > 500))
````

**Vctorize a python function**


---


If we have a function and need to pass the array elements to the function, normally we use loops to do that. If the elements are in numpy array we can simply vectorize the entire function and simply pass the array to the function. To convert a function to a vectorized function we use:


```
vectorized_def = np.vectorize(def_name)
```
vectorized_def can perform the same function in vectorized operation.



````python
# A normal function to get discounted cost

def discount_cost(rate, cost):
    """Apply a 10% discount if the rating is 4.0 or higher."""
    if rate >= 4.0:
        return cost * 0.9
    else:
        return cost
````

````python
# Getting the rating and costs from the sample numeric_data initialized before
rating = numeric_data[:, 0]
cost = numeric_data[:, 1]
````

````python
# Vectorizeing the discount function
vectorized_discount_cost = np.vectorize(discount_cost)
````

````python
# Performing vectorized function operation
vectorized_discount_cost(rating, cost)
````
